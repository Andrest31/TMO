Импорт всех нужных библиотек

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chisquare, kstest, t
```

Задача А

```python
data_A = np.array([12, 11, 10, 9, 7, 9, 8, 9, 6])
```

1) Доверительная вероятность оценки среднего с точностью e = 1

```python
n = len(data_A)
mean_A = np.mean(data_A)
std_A = np.std(data_A, ddof=1)
se = std_A / np.sqrt(n)

# Находим доверительную вероятность
alpha = 2 * (1 - stats.norm.cdf(1 / se))
print(f"Доверительная вероятность: {1 - alpha:.4f}")
```

2) Необходимое число измерений для e = 0.25, P=0.99

```python
z = stats.norm.ppf(0.995)
n_required = (z * std_A / 0.25) ** 2
print(f"Необходимое число измерений: {np.ceil(n_required)}")
```

Задача B

```python
data_B = np.array([
    102, 107, 99, 113, 96, 108, 104, 107, 100, 105, 110, 114, 100, 110, 117, 109, 117, 94, 116, 107,
    110, 95, 122, 122, 115, 102, 116, 119, 116, 118, 115, 118, 106, 103, 116, 110, 109, 121, 123, 97,
    113, 110, 119, 107, 93, 104, 115, 101, 121, 111, 107, 123, 109, 120, 100, 111, 110, 109, 106, 119,
    105, 110, 123, 106, 95, 107, 105, 118, 114, 109, 100, 120, 107, 118, 110, 99, 117, 110, 104, 113,
    108, 113, 105, 110, 117, 116, 111, 104, 115, 110, 98, 104, 110, 117, 113, 114, 107, 113, 107, 101
])
```

1) Выборочные характеристики

```python
mean_B = np.mean(data_B)
var_B = np.var(data_B, ddof=1)
std_B = np.std(data_B, ddof=1)
range_B = np.ptp(data_B)
min_B, max_B = np.min(data_B), np.max(data_B)

print(f"Среднее: {mean_B:.2f}, Дисперсия: {var_B:.2f}, СКО: {std_B:.2f}, Размах: {range_B}, Min: {min_B}, Max: {max_B}")
```

2) Доверительный интервал для математического ожидания

```python
for conf in [0.95, 0.99, 0.999]:
    ci = stats.t.interval(conf, df=len(data_B)-1, loc=mean_B, scale=std_B/np.sqrt(len(data_B)))
    print(f"Доверительный интервал при P={conf:.3f}: ({ci[0]:.2f}; {ci[1]:.2f})")
```

3) Гистограмма частот

```python
plt.hist(data_B, bins=10, alpha=0.7, color="blue", edgecolor="black")
plt.title("Гистограмма частот")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.show()
```

4) Плотность нормального распределения

```python
x = np.linspace(min_B, max_B, 100)
pdf = stats.norm.pdf(x, mean_B, std_B)

plt.plot(x, pdf, label="Нормальное распределение", color="red")
sns.histplot(data_B, bins=10, kde=True, stat="density")
plt.legend()
plt.show()
```

5) Диаграмма размаха

```python
plt.boxplot(data_B, vert=False)
plt.title("Диаграмма размаха")
plt.show()
```

6) Критерий Пирсона χ²

```python
observed_freq, bins = np.histogram(data_B, bins=10)  

# Ожидаемые частоты вычисляем по равномерному распределению выборки
expected_freq = np.ones_like(observed_freq) * np.mean(observed_freq)  

# Применяем критерий χ²
chi2, p_value = chisquare(observed_freq, expected_freq)

print(f"χ² = {chi2:.2f}, p-value = {p_value:.4f}")

if p_value > 0.05:
    print("Нет оснований отвергнуть гипотезу о нормальном распределении.")
else:
    print("Гипотеза о нормальном распределении отвергается.")
```
